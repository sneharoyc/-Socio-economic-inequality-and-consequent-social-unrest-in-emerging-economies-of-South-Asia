```python
import pandas as pd
import os

import numpy as np
from matplotlib import pyplot as plt



os.chdir("C:/Users/roych/Downloads")
data1=pd.read_csv("C:\\Users\\roych\\Downloads\\text1.csv",encoding='unicode_escape')
data2=pd.read_csv("C:\\Users\\roych\\Downloads\\text2.csv",encoding='unicode_escape')
data3=pd.read_csv("C:\\Users\\roych\\Downloads\\deprivation.csv",encoding='unicode_escape')

#Data Integration

data=pd.concat([data1, data2],ignore_index=True)


#Data cleaning

# Remove the columns
city = data.drop(columns=['Start Date','End Date','Response Type','Progress','Duration (in seconds)','Finished','Recorded Date','Response ID','User Language','Agree to survey?','How many cars on the street did you count?','Please answer the following 2 questions - Display Order','Please answer the following 3 questions - Display Order'], axis=1)
data3
```

```python
import matplotlib.pyplot as plt

df = data3.iloc[:,[1,25]].sort_values('IDACI Score')

# Fig size
#fig.set_size_inches(19,19)    
ax = df.plot.barh(x='Regions', y='IDACI Score',color=['mistyrose','pink','lightsalmon','darksalmon','salmon','palevioletred','indianred','crimson','#CD0000','#800000'])
ax.bar_label(ax.containers[0])

plt.ylabel('Least deprived                     Most deprived',fontweight='bold')
plt.xlabel('Proportion')
plt.title('Income Deprivation Affecting Children Index(IDACI) score (2019)')
leg=plt.legend()
ax.get_legend().remove()
```
